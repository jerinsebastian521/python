```
n=int(input("Enter a number:"))
f=1
if n<0:
  printf("no factorial")
elif n==0:
  printf("Factorial of 0 is 1")
else:
  for i in range (1,n+1):
     f=f*i
  print("Factorial of ",n ,"is:",f)
```

```
num = int(input("Enter how many terms: "))
 
first = 0
second = 1
 
print("Fibonacci series is:")
print(first, ",", second, end=", ")
 
for i in range(2, num):
	next = first + second
	print(next, end=", ")
 
	first = second
	second = next
```

```
def pattern(n):
  x=1
  for i in range(1, n+1):
    for j in range(1, i+1):
      print(x, end=" ")
      x=x+i
    x=i+1
    print("\n")
 
n=int(input("enter n: "))
pattern(n)
```

```
lst=input('Enter a word:')
  
freq = {} 
  
for i in lst: 
    if i in freq: 
        freq[i] += 1
    else: 
        freq[i] = 1
  
print ("Count of all characters is :" +  str(freq)) 
```

```
def print_factors(x):
   print("The factors of",x,"are:")
   for i in range(1, x + 1):
       if x % i == 0:
           print(i)


num = int(input("Enter number: "))
print_factors(num)
```

```
def get_area(a,b,c):
  
  s = (a+b+c)/2
  t=lambda s,a,b,c: (s*(s-a)*(s-b)*(s-c)) ** 0.5
  print("area of triangle:",t(s,a,b,c) )

n1=int(input("first side:"))
n2=int(input("second side:"))
n3=int(input("third side:"))
get_area(n1,n2,n3)
```
